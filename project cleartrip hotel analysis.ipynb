# data visualization part

```python
ax=sns.countplot(x='hotel_star_rating',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

# from the above graph we can conclude that the no. of count of 3 star hotels is more 

```python
ax=sns.countplot(x='state',hue='hotel_star_rating',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
sns.set(rc={'figure.figsize':()})


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
hotel_price= df.groupby(['hotel_star_rating'],as_index=False)['room_count'].sum().sort_values(by='room_count',ascending=False)
ax=sns.barplot(x='hotel_star_rating',y='room_count',data=df)


for bars in ax.containers:
    ax.bar_label(bars)
```

# from the above graph we can conclude that the no. of room_count of 5 star hotels is more

```python
hotel_price= df.groupby(['state'],as_index=False)['room_count'].sum().sort_values(by='room_count',ascending=False).head()
ax=sns.barplot(x='state',y='room_count',data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
sns.set(rc={'figure.figsize':(15,5)})

for bars in ax.containers:
    ax.bar_label(bars)

```

# from the above graph we can conclude that the no. of room_count is more in Manipur
